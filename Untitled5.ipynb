```python
import pandas as pd
import os

dataset_folder = rf"dataset"

all_data = []

for file in os.listdir(dataset_folder):
    if file.endswith(".csv"):
        dataset_path = os.path.join(dataset_folder, file)

        df = pd.read_csv(dataset_path)

        if "merged_col" in df.columns and "cScore" in df.columns:
            df_sampled = df[["merged_col", "cScore"]].sample(n=min(8000, len(df)))
            all_data.append(df_sampled)
            print(f"✅ Loaded: {file} ({len(df_sampled)} rows)")
        else:
            print(f"⚠️ Skipping {file}: Missing required columns.")

df = pd.concat(all_data, ignore_index=True)
```

```python
!pip install simpletransformers

import pandas as pd
from sklearn.model_selection import train_test_split

data = df

print(data.info())
print(data['cScore'].value_counts())

train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

train_df = pd.DataFrame({
    'text': train_data['merged_col'],
    'labels': train_data['cScore']
})

val_df = pd.DataFrame({
    'text': val_data['merged_col'],
    'labels': val_data['cScore']
})
```

```python
import re

def clean_text(text):

    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.strip()

    return text

train_df['text'] = train_df['text'].apply(clean_text)
val_df['text'] = val_df['text'].apply(clean_text)

print(train_df.head())
```

```python
from simpletransformers.classification import ClassificationModel

bert_model = ClassificationModel('bert', 'bert-base-uncased', num_labels=3, use_cuda=True)

roberta_model = ClassificationModel('roberta', 'roberta-base', num_labels=3, use_cuda=True)
```

```python
print(train_df.shape)
print(val_df.shape)
```

```python
label_mapping = {"low": 0, "medium": 1, "high": 2}
train_df["labels"] = train_df["labels"].map(label_mapping)
val_df["labels"] = val_df["labels"].map(label_mapping)

train_df = train_df.dropna(subset=["labels"])
val_df = val_df.dropna(subset=["labels"])

train_df["labels"] = train_df["labels"].astype(int)
val_df["labels"] = val_df["labels"].astype(int)
```

```python
bert_model.train_model(train_df)
```

```python
roberta_model.args.overwrite_output_dir = True
roberta_model.train_model(train_df)
```

```python
from simpletransformers.classification import ClassificationArgs

model_args = ClassificationArgs(
    num_train_epochs=3,
    train_batch_size=8,
    eval_batch_size=8,
    learning_rate=3e-5,
    max_seq_length=128,
    weight_decay=0.1,
    warmup_steps=0,
    logging_steps=50,
    save_steps=200,
)

bert_model = ClassificationModel('bert', 'bert-base-uncased', num_labels=3, args=model_args, use_cuda=True)
bert_model.args.overwrite_output_dir = True
bert_model.train_model(train_df)

roberta_model = ClassificationModel('roberta', 'roberta-base', num_labels=3, args=model_args, use_cuda=True)
roberta_model.args.overwrite_output_dir = True
roberta_model.train_model(train_df)
```

```python
result_bert, model_outputs_bert, wrong_predictions_bert = bert_model.eval_model(val_df)

print("BERT Evaluation Results:")
print(result_bert)
```

```python
result_roberta, model_outputs_roberta, wrong_predictions_roberta = roberta_model.eval_model(val_df)

print("RoBERTa Evaluation Results:")
print(result_roberta)
```

```python
from transformers import RobertaForSequenceClassification, RobertaTokenizer
import torch

model = RobertaForSequenceClassification.from_pretrained("roberta_best_model")

tokenizer = RobertaTokenizer.from_pretrained("roberta_best_model")

text = "Away from all the hustle of 96 Guntha Agricultural Land is available for Rent in Vadgaon Mulshi, Pune for 50,000"

inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

model.eval()

with torch.no_grad():
    outputs = model(**inputs)

predictions = torch.argmax(outputs.logits, dim=1).cpu().numpy()

label_map = {0: "low", 1: "medium", 2: "high"}
predicted_labels = [label_map[pred] for pred in predictions]

print(predicted_labels)
```
